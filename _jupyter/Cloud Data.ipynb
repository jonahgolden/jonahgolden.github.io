#### Downloading and cleaning
First, we will get one month of data into the form we want, then do it for all 12 months.

The ISCCP cloud coverage data has a resolution of $2.5^{\circ}$, meaning that each data point represents a region that is $2.5^{\circ}$ latitude by $2.5^{\circ}$ longitude.  Thus:  
* $180^{\circ}$ of latitudes are broken into  $\frac{180^{\circ}}{2.5^{\circ}} = 72$ discrete segments.  
* $360^{\circ}$ of longitudes are broken into  $\frac{360^{\circ}}{2.5^{\circ}} = 144$ discrete segments.
* There are $72\cdot144=10,368$ total data points per month.

First, I will access the global January cloud coverage means, and get it into a 2D array where each row represents $2.5^{\circ}$ of latitude, and each value in that row represents $2.5^{\circ}$ of longitude.

```python
import pandas as pd

def get_month_means(month):
    url = ("https://isccp.giss.nasa.gov/cgi-bin/browse_d2t2.cgi?variable=8&range=" +
           str(month) + "&yymm=&data=5&button=Download")
    month_means = pd.read_csv(url, header = None, delimiter=r"\s+")
    return month_means

january_means = get_month_means(1)
print(january_means.shape)
```

The first 144 values represent the "row" of latitudes between $88.5^{\circ}N$ and $90^{\circ}N$.  Thus, the first value represents the upper left square, which is in the range $88.5^{\circ}N - 90^{\circ}N$ and $178.5^{\circ}W - 180^{\circ}W$.

```python
num_values = 10368  # len(month_means)
resolution = 2.5  # resolution in degrees
num_lats = int(180/resolution)  # 72 = num_values/num_longs
num_longs = int(360/resolution)  # 144 = num_values/num_lats


def format_month_means(month_means):
    '''Format read_csv output into grid representing lats and longs'''
    month_means = month_means.transpose()[0]
    means_array = []
    
    for lat in range(num_lats):
        this_lat = []
        start_index = lat*num_longs
        for long in range(num_longs):
            this_lat.append(month_means[start_index+long])
        means_array.append(this_lat)
    return means_array
```

```python
formatted_january_means = format_month_means(january_means)
print("rows: " + str(len(formatted_january_means)))
print("per row: " + str(len(formatted_january_means[0])))
```

```python
# first 144 values are bottom row, left to right
# 0, 0 = bottom left
# 0, 143 = bottom right
# 71, 0 = top left
# 71, 143 = top right

def get_mean_cov(lat, long):
    lat_i = int((lat+90)/resolution)
    long_i = int((long+180)/resolution)
    return formatted_january_means[lat_i][long_i]
```

### Get and format cloud coverage data for all months

```python
def get_all_month_means():
    means = {}
    for month in range(1, 13, 1):
        month_means = get_month_means(month)
        means[month] = format_month_means(month_means)
    return means

cloud_coverage_means = get_all_month_means()
```

```python
def get_mean_cloud_coverage(lat, long, month):
    lat_i = int(lat/resolution)
    long_i = int(long/resolution)
    return cloud_coverage_means[month][-lat_i][long_i]

get_mean_cloud_coverage(179, 0, 1)
```
